```python
import tensorflow as tf
import numpy as np
import os,glob,cv2
import sys,argparse
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

tamanoDeImagenes=128
num_channels=3
imagenes = []
rutaDeDatosDeEntrenamiento=os.path.join('dataset/test/*.jpg')
listaDeArchivos = glob.glob(rutaDeDatosDeEntrenamiento)

#print("# Imágenes para prueba: "+str(len(listaDeArchivos)))
for archivo in listaDeArchivos[0:10]:
    print("Nombre del archivo "+str(archivo))
    imagen = cv2.imread(archivo)
    
    if imagen is not None:
        imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2BGR)
        imagen = cv2.resize(imagen, (tamanoDeImagenes, tamanoDeImagenes),0,0, cv2.INTER_LINEAR)
        imagen = imagen.astype(np.float64)
        imagen = np.multiply(imagen, 1.0 / 255.0)
        imagenes.append(imagen)

imagen = cv2.imread('dataset/test/Melanoma.jpg')
imagen = cv2.cvtColor(imagen, cv2.COLOR_RGB2BGR)
imagen = cv2.resize(imagen, (tamanoDeImagenes, tamanoDeImagenes),0,0, cv2.INTER_LINEAR)
imagen = imagen.astype(np.float64)
imagen = np.multiply(imagen, 1.0 / 255.0)
imagenes.append(imagen)
imagenes = np.array(imagenes)        
```

```python
## Restaurar el modelo guardado 
sess = tf.Session()
# Paso-1: recrea el gráfico de red. En este paso solo se crea el gráfico.
saver = tf.train.import_meta_graph('./model/model.meta.meta')
# Paso 2: Ahora carguemos los pesos guardados usando el método de restauración.
saver.restore(sess, tf.train.latest_checkpoint('./model'))

# Accediendo al gráfico predeterminado que hemos restaurado
graph = tf.get_default_graph()

# Ahora, veamos el op que podemos procesar para obtener la salida.
# En la red original y_pred es el tensor que es la predicción de la red
prediccionDeProbabilidadPorClase = graph.get_tensor_by_name("prediccionDeProbabilidadPorClase:0")

## Vamos a alimentar las imágenes a los marcadores de posición de entrada
tensorDeEntrada = graph.get_tensor_by_name("tensorDeEntrada:0") 
tensorDeClases = graph.get_tensor_by_name("tensorDeClases:0") 
numeroDeClases = np.zeros((1, 3)) 

# print(numeroDeClases)
for archivo in imagenes:
    imagenDeEntrada = archivo.reshape(1, tamanoDeImagenes,tamanoDeImagenes,num_channels)
    
    ### Creando el feed_dict que se requiere alimentar para calcular y_pred
    alimentarImagenConClases = {tensorDeEntrada: imagenDeEntrada, tensorDeClases: numeroDeClases}
    result=sess.run(prediccionDeProbabilidadPorClase, feed_dict=alimentarImagenConClases)
    # el resultado es de este formato [probabiliy_of_rose probability_of_sunflower]
    resultados = result[0]
    plt.imshow(archivo)
    plt.show()
    
    #print(str(resultados[2]))
    
    if(resultados[0] > resultados[1] and resultados[0] > resultados[2]):
        print("Dermatitis "+str(resultados[0]))
    
    if(resultados[1]>resultados[0] and resultados[1] > resultados[2]):
        print("Melanoma "+str(resultados[1]))
    
    if(resultados[2]>resultados[0] and resultados[2] > resultados[1]):
    #else:
        print("Seborrheic Keratosis "+str(resultados[2]))
    
```
